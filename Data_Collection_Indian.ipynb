```python
from google.colab import drive
drive.mount('/content/drive')

import wikipedia
import os
import re

# Set Wikipedia language to Malay
wikipedia.set_lang("ms")

# Base folder: Indian Festivals
base_folder = "/content/drive/MyDrive/Colab Notebooks/Indian_Festivals"
os.makedirs(base_folder, exist_ok=True)
print("Base folder created:", base_folder)

# Define festival keywords
keywords = [
    "Deepavali",
    "Thaipusam",
    "Ponggal",
    "Navaratri",
    "Holi",
    "Krishna Janmashtami",
    "Ganesh Chaturthi",
    "Onam",
    "Chitra Pournami",
    "Aadi Perukku",
    "Karthigai Deepam",
    "Hanuman Jayanti",
    "Rama Navami",
    "Vaisakhi"
]


# Sanitize file/folder names
def sanitize_filename(title):
    return re.sub(r'[\\/*?:"<>|]', "", title).replace(" ", "_")

# Main loop for each festival keyword
total_saved = 0
for keyword in keywords:
    try:
        results = wikipedia.search(keyword)
        if not results:
            print(f"No results for: {keyword}")
            continue

        # Create subfolder for this festival
        subfolder = os.path.join(base_folder, sanitize_filename(keyword))
        os.makedirs(subfolder, exist_ok=True)

        count = 0
        for title in results[:4]:  # Limit to first 3 articles
            try:
                page = wikipedia.page(title)
                filename = os.path.join(subfolder, f"{sanitize_filename(page.title)}.txt")
                with open(filename, "w", encoding="utf-8") as f:
                    f.write(page.content)
                print(f"‚úÖ Saved: {filename}")
                count += 1
                total_saved += 1
            except wikipedia.exceptions.PageError:
                print(f"‚ùå Page not found: {title}")
            except wikipedia.exceptions.DisambiguationError as e:
                print(f"‚ö†Ô∏è Disambiguation error for: {title} - options: {e.options}")
            except Exception as e:
                print(f"‚ùó Error saving '{title}': {e}")

        if count == 0:
            print(f"‚ö†Ô∏è No articles saved for: {keyword}")
    except Exception as e:
        print(f"‚ùó Error searching for '{keyword}': {e}")

print(f"\nüéâ Finished. Total articles saved: {total_saved}")
```
